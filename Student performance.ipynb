```python
# one hot encoding for nominal attribute  
columns=['Mjob','Fjob','reason','guardian']
df_encoded=pd.get_dummies(df1,columns=columns)
```

```python
df_encoded.info()
```

```python
sns.boxplot(df_encoded)
```

```python
for col in df_encoded.columns:
    plt.figure(figsize=(4, 2))
    sns.boxplot(y=df_encoded[col])
    plt.title(col)
    plt.tight_layout()
    plt.show()
```

Boxplot analysis indicated minimal presence of extreme values across variables.
Observations flagged as outliers, particularly in the failures attribute, represent valid and informative cases rather than data errors.
Hence, no outlier removal was performed to avoid loss of meaningful information.

```python
X=df_encoded.drop(columns=['G1','G2','G3'],axis=1)
```

```python
y=df_encoded['G3']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=402)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score,root_mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

Linear regression exhibited limited explanatory power due to the presence of non-linear and interaction effects among predictors.
Consequently, tree-based models were employed to better capture complex relationships, resulting in improved predictive performance.

```python
# comparing the coefficients
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])

pipe.fit(X_train, y_train)
```

```python
coeff=pipe.named_steps['lr'].coef_
features=X_train.columns
coef_df = pd.DataFrame({
    'feature': features,
    'std_coef': coeff
}).sort_values(by='std_coef', key=abs, ascending=False)
```

```python
X = df_encoded.drop(columns=['G1','G2','G3'])
X=X.astype(float)
```

```python
import statsmodels.api as sm

X_const = sm.add_constant(X, has_constant='add')
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

vif_df = pd.DataFrame()
vif_df['feature'] = X_const.columns
vif_df['VIF'] = [
    variance_inflation_factor(X_const.values, i)
    for i in range(X_const.shape[1])
]

vif_df.sort_values('VIF', ascending=False)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_df = pd.DataFrame()
vif_df['feature'] = X_const.columns
vif_df['VIF'] = [
    variance_inflation_factor(X_const.values, i)
    for i in range(X_const.shape[1])
]

vif_df.sort_values('VIF', ascending=False)
```

```python
X=df_encoded.drop(columns=['G1','G2','G3'])
y=df_encoded['G3']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=402)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=5,min_samples_leaf=10,random_state=402)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

A single decision tree exhibited severe overfitting, resulting in poor generalization performance. Consequently, ensemble-based tree methods were employed to reduce variance and improve predictive stability.

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(max_depth=20,min_samples_leaf=5)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

```python
importance=model.feature_importances_
fi=pd.DataFrame({
    'feature_name':X.columns,
    'Importance':importance}).sort_values(by='Importance',ascending=False)
```

```python
fi
```

```python
#permutation importances
from sklearn.inspection import permutation_importance
result=permutation_importance(
    model,
    X_test,
    y_test,
    n_repeats=10,
    random_state=402,
    scoring='r2'
)
perm_importance=pd.DataFrame(
    {
        'Features':X_test.columns,
        'per_importance':result.importances_mean
    }
).sort_values(by='per_importance',ascending=False)
perm_importance
```
